```python
# Vic Savage 100556060
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt


#labeling
plt.title('Highlighted points (100559637)') 
plt.axis([-10, 10, -10, 10])

#declaring my variables 
x1=[]
y1=[]
x2=[]
y2=[]
#i need to open my file with the data so python can use it  
file=open('xy_points.dat','r')

#x axis 
line=file.readline() #i need the programe to them read the opened file
x=line.split() 
x=np.array(x) #here i had the array contaian all the x values in it

#y axis
line=file.readline()
y=line.split()
y=np.array(y) #here i had the array contaian all the y values in it

#the blue needs to be x<-2 and y>=-5 and y<=0
for i in range(len(x)):
    if float(x[i])<-2 and float(y[i])>=-5.0 and float(y[i])<=0.0:
        x1.append(x[i])
        y1.append(y[i]) 
        #This will give us the changed colours         


#The green needs to be a circle with radius(0,5)
for i in range(len(x)):
    if (float(x[i])**2+(float(y[i])-5)**2)<25: #the equation of a circle but needs to be within 25 
        x2.append(x[i])
        y2.append(y[i]) 
         #This will give us the changed colours         

plt.plot(x,y,'ro') #red dots
plt.plot(x1,y1,'bo') #blue dots
plt.plot(x2,y2,'go') #green dots

plt.show() 
```
