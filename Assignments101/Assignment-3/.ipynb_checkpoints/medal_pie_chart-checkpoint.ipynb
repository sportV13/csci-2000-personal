```python
#Vic Savage 100556060
from pylab import *
%matplotlib inline
import numpy as np
import matplotlib.pyplot as mplt


#the medals were too small to see so i needed to make the pie chart larger
figure(1, figsize=(10,10))

ax = axes([0.1, 0.1, 0.8, 0.8])
#my labels for the chart 
mplt.suptitle("2012 Olympic Gold Medals (100556060)")
labels = 'USA', 'CHIN', 'GBR', 'RUS', 'KOR', 'GER', 'FRA', 'ITA', 'HUN', 'AUS', 'OTHER'
medals = [46, 38, 29, 24, 13, 11, 11, 8, 8, 7, 107]
total=sum(medals)
#to make my chart colourfull i used this and also looked up the index for the colours
colours = ["#4444ff", "#ff4500", "#ff1493", "#663300", "#8a2be2", "#ffff00", "#663300", "#006400", "#330033", "#CC9999", "#660000"]

#formats all my variables into the chart
mplt.pie(medals, explode=explode, labels=labels, autopct=lambda(p): '{:.0f}'.format(p * total / 100), colors=colours, shadow=True, startangle=90)


mplt.show()
```
